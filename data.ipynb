```python
import pandas as pd


df = pd.read_csv('financial_status.csv')

print("Dataset shape:", df.shape)
print("Column names:")
print(df.columns.tolist())

df.head()
```

```python
import json
import os


os.makedirs('jsons', exist_ok=True)
financial_data = df.to_dict('records')


with open('jsons/financial_data.json', 'w') as f:
    json.dump(financial_data, f, indent=2, default=str)

print(f"Successfully saved {len(financial_data)} records to jsons/financial_data.json")
print(f"Sample record structure:")
print(json.dumps(financial_data[0], indent=2, default=str))
```
